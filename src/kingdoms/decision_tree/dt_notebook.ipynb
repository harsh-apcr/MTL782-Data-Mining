```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import KFold

from sklearn import metrics
from sklearn.tree import plot_tree


import matplotlib.pyplot as plt

path = '../../codon-usage-data-set/'
df = pd.read_csv(path + 'codon_usage.csv', low_memory = False)

df = df[df.Kingdom != 'plm']

df = df[pd.to_numeric(df['UUU'], errors='coerce').notnull()]
df = df[pd.to_numeric(df['UUC'], errors='coerce').notnull()]

df.UUU = df.UUU.astype('float64')
df.UUC = df.UUC.astype('float64')

df.Kingdom = df.Kingdom.astype('category')

df = df[df.Ncodons > 1000]

df['Kingdom'] = df['Kingdom'].replace(
                      {'pln' : 'euk' ,
                       'inv' : 'euk' ,
                       'vrt' : 'euk' ,
                       'mam' : 'euk' ,
                       'rod' : 'euk' ,
                       'pri' : 'euk' 
                      })

df = df.reset_index(drop=True)
df.groupby('Kingdom').size()
```

```python
# Tuning the hyper-parameter (max_leaf_nodes of the decision_tree_classifier)
from dt_training import tree_training

codon_columns = [c for c in df.columns if c not in ['Kingdom', 'DNAtype', 'SpeciesID', 'Ncodons', 'SpeciesName']]

kfold = KFold(5, shuffle = True, random_state = 1)

print('Iteration Starts')
for i in range(2,20):
    tot_train = 0
    tot_test = 0
    for idx_train, idx_test in kfold.split(df):
        X_train = df.loc[idx_train, codon_columns]
        y_train = df.loc[idx_train, 'Kingdom']

        X_test = df.loc[idx_test, codon_columns]
        y_test = df.loc[idx_test, 'Kingdom']

        train, test = tree_training(i,X_train, y_train, X_test, y_test)
        tot_train += train
        tot_test += test
    
    # print the average training accuracy for model with max_leaf_nodes = i
    print(f'max_leaf_nodes:{i} ,training-accuracy:{round(tot_train / 5, 3)}, test-accuracy:{round(tot_test / 5, 3)}')
print('Iteration Ends\n----------------------------')
    
```

We can clearly observe from above outcomes , best performance on validation set (with minimum number of leaf nodes) is for 11 nodes 

`max_leaf_nodes:11 ,training-accuracy:0.617, test-accuracy:0.613`

Now, we'll create our final model using `max_leaf_nodes = 11`

```python
model_tree = DecisionTreeClassifier(max_leaf_nodes = 11, class_weight = 'balanced')
model_tree.fit(df.loc[:,codon_columns], df['Kingdom'])
```

Print the final tree

```python
plt.figure(figsize = (20,10))

plot_tree(model_tree,
         feature_names = codon_columns,
         class_names = ['arc', 'bct', 'phg', 'euk', 'vrl'],
         rounded = True,
         filled = True)

plt.show()
```
